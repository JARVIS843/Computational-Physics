```python
from math import *
import matplotlib.pyplot as plt
import numpy as np


def Air_Drag(theta,V0 = 100 , b = 0.5 , m=100):
    V0 = 100
    b = 0.5
    m = 100
    g = 9.8 ; dt = 0.01
    theta = np.deg2rad(theta)
    r0 = np.array([0.0,0.0])
    Vinit = np.array([V0 * np.cos(theta) , V0 * np.sin(theta)])
    r = r0
    v= Vinit
    r_values = np.copy(r0)
    while r[1] >= 0:
        a = - b * np.linalg.norm(v)*v/m + np.array([0,-g])
        v += a*dt
        r += v*dt
        r_values = np.vstack((r_values,r)) ; plt.axis('equal')
        
    
    plt.plot(r_values[:,0] , r_values[: , 1] , 'r') ; plt.grid() ; plt.axhline() ; plt.axvline()
    return (r_values[-1:,0][0] , theta)
```

```python
Air_Drag(1)
```

```python
def MaximusDis(theta , NumIteration):
    theta = np.linspace(0,theta,NumIteration)
    dis = list(map(Air_Drag, theta))
    print(f"The Max Distance is {np.max(dis)} with an angle of {theta[dis.index(np.max(dis))]}")
```

```python
MaximusDis(90,100)
```
