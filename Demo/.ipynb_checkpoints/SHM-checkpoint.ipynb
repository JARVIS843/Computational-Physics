```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

def SHM_Euler(Omega0 , X0 , V0):
    t = np.linspace(0,20*pi/Omega0,1000)
    dt = t[1] - t[0]
    V = np.zeros(len(t)) ; V[0] = V0
    X = np.zeros(len(t)) ; X[0] = X0
    for i in range(1,len(V)):
        V[i] = V[i-1] - Omega0**2* X[i-1]*dt
        X[i] = X[i-1] + V[i-1] * dt
    plt.plot(t,V,'r') ; plt.plot(t,X,'g')
```

```python
SHM_Euler(2,10,10)
```

```python
def SHM_EC(Omega0 , X0 , V0):
    t = np.linspace(0,20*pi/Omega0,1000)
    dt = t[1] - t[0]
    V = np.zeros(len(t)) ; V[0] = V0
    X = np.zeros(len(t)) ; X[0] = X0
    for i in range(1,len(V)):
        V[i] = V[i-1] - Omega0**2* X[i-1]*dt
        X[i] = X[i-1] + V[i] * dt
    plt.plot(t,X,'g')
```

```python
SHM_EC(2,10,10)
```
