```python
import matplotlib.pyplot as plt
from math import*
import numpy as np
```

```python
def Radioactivity(N0 , r , SampleSize = 50):
    t = np.linspace(0,5/r,SampleSize)
    dt = t[1] - t[0]
    N = np.zeros(len(t)) ; N[0] = N0
    for i in range(1,len(N)):
        N[i] = N[i-1] * (1-r*dt)
    plt.plot(t,N0*np.exp(-r*t) , 'o') #A
    plt.plot(t,N , 'o') #Prediction
    plt.xlabel("T (s)") ; plt.ylabel("Num of undecayed Atoms (atoms)"); plt.grid() ; plt.title('Radioactive Decay')
    plt.axvline(color='r');plt.axhline(color='r')
```

```python
Radioactivity(1000, 5 , 10)
```

```python
def RadioActivity_2(Na0 , Nb0 , Ra, Rb , SampleSize = 50):
    t = np.linspace(0,5/min(Ra,Rb),SampleSize)
    dt = t[1] - t[0]
    Na = np.zeros(len(t)) ; Na[0] = Na0
    Nb = np.zeros(len(t)) ; Nb[0] = Nb0
    for i in range(1,len(Na)):
        Na[i] = Na[i-1] - Ra*Na[i-1] * dt
        Nb[i] = Nb[i-1] * (1-Rb *dt) + Ra * dt * Na[i-1]
    plt.plot(t,Na,'ro')
    plt.plot(t,Na0 * e**(-Ra*t),'g')
    plt.plot(t,Nb,'o')
    plt.plot(t,e**(-Rb*t)*(Nb0 + Ra*Na0*((e**((Rb-Ra)*t)-1)/(Rb-Ra))))
    
    
    plt.legend(['Predicted Na','Actual Na','Predicted Nb','Actual Nb'])
    plt.xlabel("T (s)") ; plt.ylabel("Num of undecayed Atoms (atoms)"); plt.grid() ; plt.title('Radioactive Decay')
    plt.axvline(color='r');plt.axhline(color='r')
```

```python
RadioActivity_2(1000,1000,5,6)
```
