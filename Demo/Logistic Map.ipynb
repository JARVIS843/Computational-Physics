```python
import matplotlib.pyplot as plt
from math import *
import numpy as np

def EXP_Growth(r , X0 , N , End):
    x_axis = np.linspace(X0 , End , 1000)
    X = np.zeros(len(x_axis))
    X[0] = X0
    for iteration in range(1,N + 1):
        X[iteration] = r * X[iteration -1] * (1-X[iteration-1])
    plt.plot(x_axis , X)
    print(X)
```

```python
EXP_Growth(0.001 , 10 , 100 , 20)
```
