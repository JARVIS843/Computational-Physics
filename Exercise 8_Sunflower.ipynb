```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
```

```python
def SunFlower():
    plt.axes(projection = 'polar')
    rads = np.arange(0, (2 * np.pi), 0.01)
    s = 1
    offset = 0;
    for rad in rads:
        r = sqrt(s)
        plt.plot(rad * 2.3999632297 + offset,r,'g',marker='o')
        s+=1
        offset += 2.3999632297
        
    plt.show()
    
```

```python
SunFlower()
```
