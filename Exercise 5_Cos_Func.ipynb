```python
from math import factorial
import matplotlib.pyplot as plt
from math import pi
from numpy import *
def TaylorSeriers(order = 1 , x = 0):
    total = 0
    even = 0
    for items in range(1 , order + 1):
        if items % 2 == 0:
            total -= (x**even)/factorial(even)
        else:
            total += (x**even)/factorial(even)
        even +=2
    return total
```

```python
x = linspace(0,(3*pi)/2,100)
plt.plot(x,TaylorSeriers(5,x)) ; plt.xlabel("x") ; plt.ylabel("y")
```
