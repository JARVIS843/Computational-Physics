```python
import matplotlib.pyplot as plt
from sympy import *
import numpy as np
```

## Question A:

```python
a = 3.4 * 10**(-10); e = 1.65 * 10**(-21) ; r = np.linspace(0.95 * a,2 * a,1000)

def U(r):
    U = 4*e*(((a/r)**12) - (a/r)**6)
    return U

plt.plot(r,U(r)) ; plt.xlabel("r (m)") ; plt.ylabel("U (J)") ; plt.title("Potential Energy between Two Argon Atoms")
```

## Question B

```python
r = np.linspace(1 * a,3 * a,1000)
def F(r):
    diff = (24*e*(a**6)*((r**6) - 2*(a**6)))/(r**13)
    return -diff

plt.plot(r,F(r));plt.xlabel("r (m)") ; plt.ylabel("F (N)") ; plt.title("Force between Two Argon Atoms")
```

## Question C

```python
k = 4 * e * 36 / 2 ** (1 / 3) / a ** 2 * 0.5; V0 = 4 * e * -1/4; r0 = a * 2 ** (1 / 6); r = np.linspace(6.5 * 5.2917721067121 * 10 ** -11,8 * 5.2917721067121 * 10 ** -11,1000)

def U(r):
    U = 4*e*(((a/r)**12) - (a/r)**6)
    return U

def V(r):
    v = 10000*k *(r-r0)**2 + V0
    return v

plt.plot(r,V(r),'r',r,U(r))
plt.legend(["Potential Energy", "Harmonic Function"])
```
