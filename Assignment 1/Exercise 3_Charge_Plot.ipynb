```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```python
def Charge(t = 0.1):
    q0 = 10 ; R = 60 ; L = 9 ; C = 0.00005
    charge = q0 * e**((-R*t)/(2*L)) * np.cos(t * sqrt((1/(L*C)) - (R/(2*L))**2))
    return charge
```

```python
TimeList = np.linspace(0,0.8,100)
```

```python
plt.plot(TimeList , Charge(TimeList)); plt.xlabel("Time (s)") ; plt.ylabel("Charge (c)")
```
