```python
import matplotlib.pyplot as plt
from numpy import *
```

```python
def Density(Tc = 0):
    density = 5.5289 * 10**(-8) * Tc**3 - 8.5016 * 10**(-6) * Tc**2 + 6.5622 * 10**(-5) * Tc + 0.99987
    return density
```

```python
TcList = linspace(0,100,1000)
plt.plot(TcList , Density(TcList)) ; plt.xlabel("Temperature (C)") ; plt.ylabel("Density (g/cm3)") ;plt.grid()
```
