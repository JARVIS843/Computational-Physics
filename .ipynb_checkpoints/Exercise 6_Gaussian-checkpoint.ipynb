```python
from math import *
from numpy import *
import matplotlib.pyplot as plt
```

```python
SD = (1,1.5,2)
def Gaussian(sd , x = 0):
    g = 0
    g = 1/(sd * sqrt(2*pi)) * e**(-(x**2)/(2*(sd**2)))
    return g
```

```python
for sd in range(0,3):
    listSD = list(Gaussian(SD[sd], linspace(-2,2,1000)))
    plt.plot(linspace(-2,2,1000),listSD) ; plt.xlabel("x") ; plt.ylabel("y")
```
